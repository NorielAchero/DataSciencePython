```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
student = {"Noriel": 100, "Angelo": 78, "John Michael": 75}
```

```python
names = list(student.keys())
values = list(student.values())
```

```python
plt.bar(names, values)
plt.title("Bar Plot of Grades")
plt.xlabel("Student")
plt.ylabel("Grades")
plt.grid(True)
plt.show()
```

```python
plt.barh(names, values, color = "red")
plt.title("Bar Plot of Grades")
plt.xlabel("Grades")
plt.ylabel("Student")
plt.grid(True)
plt.show()
```
