```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
y = 2 * x
y
```

```python
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show
```

```python
plt.plot(x,y, color = 'black', linestyle = ':', linewidth=5)
plt.show
```

```python
x2 = x * 2
y2 = 2 * x2
```

```python
plt.plot(x,y, color = 'black', linestyle = ':', linewidth=2)
plt.plot(x,y2, color = 'red', linestyle = '-', linewidth=4)
plt.title("Multiple Plots")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(True)
plt.show
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, color = 'black', linestyle = ':', linewidth=2)

plt.subplot(1,2,2)
plt.plot(x,y2, color = 'red', linestyle = '-', linewidth=4)

plt.show()
```
