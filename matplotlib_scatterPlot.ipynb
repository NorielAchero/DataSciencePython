```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(10, 100, 10)
y = np.random.randint(1, 10, 9)
z = np.random.randint(1, 10, 9)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y, marker = "*", c = "blue", s = 50)
```

```python
plt.scatter(x,y, marker = "*", c = "blue", s = 50)
plt.scatter(x,z, marker = ".", c = "red", s = 50)
```

```python
plt.subplot(2, 1, 1)
plt.scatter(x,y, marker = "*", c = "blue", s = 50)
plt.subplot(2, 1, 2)
plt.scatter(x,z, marker = ".", c = "red", s = 50)
```
